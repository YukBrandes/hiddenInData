## 1.1 什么是机器学习
机器学习的明确定义比较晦涩，举例，你正在使用某手机视频软件观看剧集，你一定会发现专门的推荐栏，并且视频总是越刷越有，好似无穷尽也；那是因为软件的后台根据用户历史浏览数据作为训练数据，将你是否点击收藏作为目标变量，预判了一个系列视频与你喜好的相关性，再将你喜欢的推送给你，这就是个机器学习任务训练及其应用过程，抽象如下：

## 1.2 机器学习的优势
示例情景，在机器学习普及以前，也可以通过编写规则的形式实现，比如说画风、作者、主题等，这种方法也适合数据积累不充分的新兴业务。但是伴随着业务的深化、复杂程度攀升，该方法的辨识度可能会迅速衰变、维护成本极高；此时，基于机器学习方法将会降低维护成本，提高识别精度和广度。

## 1.3 机器学习的分类
机器学习的分类颇有“横看成岭侧成峰”的感觉，角度不同，定义不同，主要的几种定义如下：

**根据是否具有目标变量：**
- 监督学习：建模数据包含标签。比如，示例中的视频收藏标签。
- 无监督学习：建模数据包不含标签。比如，经典的“啤酒+尿不湿”的故事。
- 半监督学习：无监督与有监督的叠加。比如，智能手机相册有一种自动识别相同的人出现在哪些相片中的功能，当你给添加上姓名后，算法就会命名每张照片上的每个人物。
- 强化学习：算法也可以称作智能体（agent）在没有既定目标的前提下，可以对环境进行观察，选择和执行动作，通过对其行为的奖惩机制给予其反馈，目标是学习最佳方法（称为策略，policy），以得到长久的最大奖励，从而决定了智能体在给定情况下应采取的行动。

**根据是否进行持续学习：**
- 批量学习：必须用所有可用数据进行训练，会占用大量时间和计算资源，所以一般线下做。
- 在线学习：数据实例持续训练，可以一次一个或几个实例，学习步骤快且廉价，可以动态地学习新数据。

**根据是否进行归纳推广：**
- 基于实例学习：即，记忆学习，利用相似度测量将已知样本结果推广到新例。
- 基于模型学习：建立样本的模型，然后使用这个模型进行预测。

以上概念可以根据实际问题组合使用，比如**基于实例的有监督学习任务**。

## 1.4 机器学习的方法论
对机器学习的概念有了一定的了解后，接下来聊聊机器学习的方法论，先看看下面的物理实验：

![](https://ss3.bdstatic.com/70cFv8Sh_Q1YnxGkpoWK1HF6hhy/it/u=818847796,283231577&fm=26&gp=0.jpg)

这张图展示了有关**牛顿第二定律**的实验，你将通过测力计测出左侧物体的总重力（$F$），通过称重计测出物体的总质量（$m$）,通过一系列计算求出右侧被牵拉物体的加速度（$a$），汇总如类似下表：

|重力 $F$|质量 $m$|加速度 $a$|
|----|----|----|
|5|5|0.99|
|10|5|1.98|
|15|5|2.97|
|20|5|3.96|
通过观察（或者绘制图像），你发现 $F$ 与 $m$、$a$ 存在相关关系，你提出猜测：$ F = ma + \theta $，基于实验迭代得出$ F = ma $为最佳解，误差是由别的因素引起的。

与之科学研究类似，机器学习也会经历假设、评价和优化三个阶段：

- **模型假设**：提出能够拟合已知样本的模型假设。
- **损失函数**：衡量预测值与真实值好坏程度的指标。
- **寻解算法**：基于已知样本寻找模型参数使损失函数值最小的过程。

其中，需要重点关注的是**模型假设**与**损失函数**。

关于模型假设，存在已经确定的模型被新的数据推翻，那么此时需要重新评估模型有效性或提出新的模型假设。

关于损失函数的设定需注意兼顾如下**偏差**与**方差**：
- **偏差**：预测值的期望与真实值之间的差距；偏差越大，越偏离真实数据。
- **方差**：预测值的变化范围，离散程度（即离期望值的距离）；方差越大，数据的分布越分散。

但是基于训练集的低偏差、低方差建模可能会带来机器学习的另一个常见问题：**过拟合**（反之则为“**欠拟合**”），即过分专著于训练集的个性，未捕捉客观规律，难有效推广。所以一个低偏差、低方差的有效模型，是基于训练集和测试集同时进行考量的结果。解决过拟合的一般方法：一是，**扩充训练数据**；二是，**正则化**。
