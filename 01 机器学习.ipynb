# 1 机器学习
## 1.1 什么是机器学习
关于数据科学理论的书籍，名家荟萃，可自行收集，本文只是为后续实践部分的开展梳理通识性的概念，以便对机器学习毫无经历者更好的理解后续内容。
机器学习的明确定义比较晦涩，所以用个例子进行“意会”。例如，你在使用某视频软件观看剧集，它可以根据你的观看历史作为训练数据，将你是否点击收藏作为目标变量，目的就是预测一个新上线的剧集是否是你喜欢的，那么这就是个机器学习任务。可以抽象如下：

## 1.2 机器学习的优势
其实针对于刚刚的情景，在机器学习普及以前，也可以通过编写规则的形式实现，比如说画风、作者、主题等，这种方法也是目前数据积累不充分业务早期所采用的。但是随着业务的积累 ，其复杂程度、维护成本都会奇高，此时，基于统计概率的机器学习方法将会大幅精简维护成本，提高识别精度和广度。
## 1.3 机器学习的分类
机器学习的分类颇有“横看成岭侧成峰”的感觉，角度不同，定义不同，主要的几种定义如下：

**根据是否具有目标变量：**
- 监督学习：建模数据包含标签。比如，刚刚例举情景中的视频收藏标签。
- 无监督学习：建模数据包不含标签。比如，大家熟知的“啤酒+尿不湿”的故事。
- 半监督学习：无监督与有监督的叠加。比如，智能手机相册种有一种自动识别相同的人出现在哪些相片中的功能，当你给添加上姓名后，算法就会命名每张照片上的每个人物。
- 强化学习：算法也可以称作智能体（agent）在没有既定目标的前提下，可以对环境进行观察，选择和执行动作，通过对其行为的奖惩机制给予其反馈，目标是学习最佳方法（称为策略，policy），以得到长久的最大奖励，从而决定了智能体在给定情况下应采取的行动。

**根据是否进行持续学习：**
- 批量学习：必须用所有可用数据进行训练，会占用大量时间和计算资源，所以一般线下做。
- 在线学习：数据实例持续训练，可以一次一个或几个实例，学习步骤快且廉价，可以动态地学习到达的新数据。

**根据是否进行归纳推广：**
- 基于实例学习：即，记忆学习，利用相似度测量将已知样本结果推广到新例。
- 基于模型学习：建立样本的模型，然后使用这个模型进行预测。

以上概念可以根据实际问题组合使用，比如**基于实例的有监督学习任务**。
## 1.4 机器学习的方法论
对机器学习的概念有了一定的了解后，接下来聊聊机器学习的方法论，我们先来看一个例子：

![](https://ss3.bdstatic.com/70cFv8Sh_Q1YnxGkpoWK1HF6hhy/it/u=818847796,283231577&fm=26&gp=0.jpg)

这张图展示了有关**牛顿第二定律**的实验，你将通过测力计测出左侧物体的总重力（$F$），通过称重计测出物体的总质量（$m$）,通过一系列计算求出右侧被牵拉物体的加速度（$a$），汇总如类似下表：

|重力 $F$|质量 $m$|加速度 $a$|
|----|----|----|
|5|5|0.99|
|10|5|1.98|
|15|5|2.97|
|20|5|3.96|
通过观察（或者绘制图像），你发现 $F$ 与 $m$、$a$ 存在相关关系，部分误差是由某些因素引起的，最后你大胆猜测并验证：$ F = ma $ 为最佳解。

与之类似，机器学习也会经历假设、评价和优化三个阶段：

- **模型假设**：提出能够拟合已知样本的模型假设（未来新的数据打破假设，可以重新提出新的模型假设）。
- **损失函数**：衡量预测值与真实值好坏程度的指标。
- **寻解算法**：基于已知样本寻找模型参数使损失函数值最小的过程。

其中，损失函数怎么设定的呢？有两个重要的指标：
- **偏差**：预测值的期望与真实值之间的差距；偏差越大，越偏离真实数据。
- **方差**：预测值的变化范围，离散程度（即离期望值的距离）；方差越大，数据的分布越分散。

但是基于训练集的低偏差、低方差建模可能会带来机器学习的另一个常见问题：**过拟合**（反之则为“欠拟合”），即过分专著训练集个性，未捕捉典型规律，难有效推广。所以一个低偏差、低方差的有效模型，是基于训练集和测试集同时进行考量的结果。解决过拟合的一般方法：一是，**扩充训练数据**；二是，**正则化**。
