# 1 机器学习
## 1.1 什么是机器学习
关于数据科学理论的书籍，名家荟萃，可自行收集，本文只是为后续实践部分的开展梳理通识性的概念，以便对机器学习毫无经历者更好的理解后续内容。
机器学习的明确定义比较晦涩，所以用个例子进行“意会”。例如，你在使用某视频软件观看剧集，它可以根据你的观看历史作为训练数据，将你是否点击收藏作为目标变量，目的就是预测一个新上线的剧集是否是你喜欢的，那么这就是个机器学习任务。可以抽象如下：

## 1.2 机器学习的优势
其实针对于刚刚的情景，在机器学习普及以前，也可以通过编写规则的形式实现，比如说画风、作者、主题等，这种方法也是目前数据积累不充分业务早期所采用的。但是随着业务的积累 ，其复杂程度、维护成本都会奇高，此时，基于统计概率的机器学习方法将会大幅精简维护成本，提高识别精度和广度。
## 1.3 机器学习的分类
机器学习的分类颇有“横看成岭侧成峰”的感觉，角度不同，定义不同，主要的几种定义如下：

**根据是否具有目标变量：**
- 有监督学习：训练数据包含标签。比如，刚刚例举情景中的视频收藏标签。
- 无监督学习：训练数据包不含标签。比如，大家熟知的“啤酒+尿不湿”的故事。
- 半监督学习：无监督与有监督的叠加。比如，智能手机相册种有一种自动识别相同的人出现在哪些相片中的功能，当你给添加上姓名后，算法就会命名每张照片上的每个人物。
- 强化学习：算法也可以称作智能体（agent）在没有既定目标的前提下，可以对环境进行观察，选择和执行动作，通过对其行为的奖惩机制给予其反馈，目标是学习最佳方法（称为策略，policy），以得到长久的最大奖励，从而决定了智能体在给定情况下应采取的行动。

**根据是否进行持续学习：**
- 批量学习：必须用所有可用数据进行训练，会占用大量时间和计算资源，所以一般线下做。
- 在线学习：数据实例持续训练，可以一次一个或几个实例，学习步骤快且廉价，可以动态地学习到达的新数据。

**根据是否进行归纳推广：**
- 基于实例学习：即，记忆学习，利用相似度测量将已知样本结果推广到新例。
- 基于模型学习：建立样本的模型，然后使用这个模型进行预测。

以上概念可以根据实际问题组合使用，比如**基于实例的有监督学习任务**。
